```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import seaborn as sns
import sys
```

```python
X, y = make_moons(n_samples=500, noise=0.05)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, shuffle=True)

df = pd.DataFrame(dict(x=X[:,0], y = X[:,1], label = y))

colors = {0: 'red', 1: 'blue'}
grouped = df.groupby('label')

fig, ax = plt.subplots()
for key, group in grouped:
    group.plot(ax = ax, kind = 'scatter', x = 'x', y = 'y', label = key, color = colors[key])
plt.show()
```

```python
import torch
from torch import nn
import torch.optim as optim
```

```python
# Vanilla Neural Network
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(2, 8),
            nn.Sigmoid(),
            nn.Linear(8, 8),
            nn.Sigmoid(),
            nn.Linear(8, 1),
            nn.Sigmoid(),
        )

    def forward(self, x):
        logits = self.model(x)

        return logits
```

```python
model = NeuralNetwork().to('cpu')
print(model)
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.2)
```

```python
# Convert test data in numpy array to torch tensors
X_train_t = torch.tensor(X_train, dtype=torch.float32, requires_grad = False)
y_train_t = torch.tensor(y_train, dtype=torch.float32, requires_grad = False)

y_train_t = y_train_t.unsqueeze(1)
```

### Batch processing

```python
# Write a simple batch generator
def batch_processor(dataset: torch.Tensor, batch_size: int) -> torch.Tensor:
    n_batches = int(len(dataset)/ batch_size)
   
    for i in range(n_batches):
        # Slice tensors into batch size
        batch = dataset[i* batch_size: (i+1) * batch_size,]
        yield batch 
```

```python
# NOTE: Future work: Apparently the dataset is shuffled prior to each epoch 
```

```python
n_epochs = 10
n_batch = 5

# Training loop
for epoch in range(n_epochs):
    
    for i in batch_processor(X_train_t, n_batch):
        # Forward pass
        y_pred = model.forward(X_train_t)

        loss = criterion(y_pred, y_train_t)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()

        # Update weights
        optimizer.step()

    # Evaluation
    print(f"Epoch {epoch}: {loss}")
```

```python
X, Y = np.mgrid[-2:3:100j,-1:2:100j]
X = X.flatten()
Y = Y.flatten()

# Visualize decision boundary 
colors = {0: 'red', 1: 'blue'}
grouped = df.groupby('label')

fig, ax = plt.subplots()
prediction = [1 if model.forward(torch.tensor([x,y], dtype=torch.float32)) > 0.8 else 0 for x, y in zip(X,Y)]
for key, group in grouped:
    group.plot(ax = ax, kind = 'scatter', x = 'x', y = 'y', label = key, color = colors[key])

ax.scatter(X, Y, alpha = 0.1, c = prediction)
plt.show()
```
